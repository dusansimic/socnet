```julia
using Graphs
using Karnak
using NetworkLayout
using Colors

g = SimpleGraph(100, 1650)

@drawsvg begin
  background("black")
  sethue("grey40")
  fontsize(8)
  drawgraph(g,
    layout=stress,
    vertexlabels = 1:nv(g),
    vertexfillcolors = 
      [RGB(rand(3)/2...)
        for i in 1:nv(g)]
  )
end
```
